```python
def load_results(path, suffix = ''):
    files = [join(path, f) for f in listdir(path) if isfile(join(path, f)) and f.endswith('.json')]
    
    res = {
        'File': [],
        'Avg Delay 1': [],
        'Avg Delay 2': [],
        'Avg Delay I': [],
        'Avg Delay AM': [],
        'Avg Delay NM': [],
        'Avg Delay MDP': [],
        'Avg M Cost AM': [],
        'Avg M Cost NM': [],
        'Avg M Cost MDP': [],
        '% time 2 better': [],
        '% Opt 1' : [],
        '% Opt 2': [],
        '% Opt AM': [],
        '% Opt NM': [],
        '% Opt MDP': []
    }

    n_states = []
    
    for file in files:
        try:
            benchmark = json.load(open(file + suffix + '.benchmark', 'r'))['results']
            series = json.load(open(file + '.series', 'r'))['series']
            models = json.load(open(file, 'r'))['models']

            for model in models:
                n_states.append(len(model['lengths']))

            pair = tuple(json.load(open(file, 'r'))['pair'])
            if pair not in whitelist:
                continue

            opt_path = np.min(np.vstack((series[0]['raw_rtt'], series[1]['raw_rtt'])), axis=0)

            res['File'].append(file)

            res['Avg Delay 1'].append(np.mean(series[0]['raw_rtt']))
            res['Avg Delay 2'].append(np.mean(series[1]['raw_rtt']))
            res['Avg Delay I'].append(np.mean(opt_path))

            res['Avg Delay AM'].append(np.mean(benchmark[0]['delay_hist']))
            res['Avg Delay NM'].append(np.mean(benchmark[1]['delay_hist']))
            res['Avg Delay MDP'].append(np.mean(benchmark[2]['delay_hist']))

            # IMPORTANT: Assume measure cost of 1
            res['Avg M Cost AM'].append(np.sum(benchmark[0]['action_hist']) / len(benchmark[0]['delay_hist']))
            res['Avg M Cost NM'].append(np.sum(benchmark[1]['action_hist']) / len(benchmark[1]['delay_hist']))
            res['Avg M Cost MDP'].append(np.sum(benchmark[2]['action_hist']) / len(benchmark[2]['delay_hist']))

            res['% time 2 better'].append(np.sum(np.array(series[0]['raw_rtt']) > np.array(series[1]['raw_rtt'])) / len(series[0]['raw_rtt']))

            res['% Opt 1'].append(np.mean(np.array(series[0]['raw_rtt']) == opt_path))
            res['% Opt 2'].append(np.mean(np.array(series[1]['raw_rtt']) == opt_path))
            res['% Opt AM'].append(np.mean(np.array(benchmark[0]['delay_hist']) == opt_path))
            res['% Opt NM'].append(np.mean(np.array(benchmark[1]['delay_hist']) == opt_path))
            res['% Opt MDP'].append(np.mean(np.array(benchmark[2]['delay_hist']) == opt_path))
        except Exception as e:
            print(e)

    df = pd.DataFrame.from_dict(res)
    df.set_index('File', inplace=True)
    
    # Measure model fit
    df['Avg AM-I'] = (df['Avg Delay AM']-df['Avg Delay I'])# / df['Avg Delay I'] * 100
    df['Avg NM-I'] = (df['Avg Delay NM']-df['Avg Delay I'])# / df['Avg Delay I'] * 100
    df['Avg MDP-I'] = (df['Avg Delay MDP']-df['Avg Delay I']) # / df['Avg Delay I'] * 100
    df['Avg MDP-AM'] = (df['Avg Delay MDP']-df['Avg Delay AM']) # / df['Avg Delay I'] * 100
    df['Avg 1-I'] = (df['Avg Delay 1']-df['Avg Delay I']) # / df['Avg Delay I'] * 100
    df['Avg 2-I'] = (df['Avg Delay 2']-df['Avg Delay I'])# / df['Avg Delay I'] * 100
    df['Avg 1-2'] = (df['Avg Delay 1']-df['Avg Delay 2'])

    df['Avg Penalized AM-I'] = (df['Avg Delay AM']+df['Avg M Cost AM']-df['Avg Delay I'])# / df['Avg Delay I'] * 100
    df['Avg Penalized NM-I'] = (df['Avg Delay NM']+df['Avg M Cost NM']-df['Avg Delay I']) #/ df['Avg Delay I'] * 100
    df['Avg Penalized MDP-I'] = (df['Avg Delay MDP']+df['Avg M Cost MDP']-df['Avg Delay I']) #/ df['Avg Delay I'] * 100
    
    return n_states, df
```
